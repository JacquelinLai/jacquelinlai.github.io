```python
data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
data = pd.read_csv(data_url)
df = data[data["Year Acquired"] > 1975]
```

```python
df = pd.DataFrame({
    'Year Acquired': list(set(df["Year Acquired"].to_numpy())),
    'Square Footage': df["Square Footage"].groupby(df["Year Acquired"]).sum().to_numpy()
})

chart = alt.Chart(df).mark_line().encode(
     x='Year Acquired',
     y='Square Footage'
)
chart
```

```python
import pandas as pd

selected_agencies = ['Department of Natural Resources', 'Department of Transportation']
df_2 = data[data['Agency Name'].isin(selected_agencies)]


grouped_data = df_2.groupby(["Agency Name", "Bldg Status"])

aggregated_data = grouped_data["Bldg Status"].count().reset_index(name="count")

aggregated_data
```

```python
import pandas as pd
from altair import Chart

chart = Chart(aggregated_data)

# Encode x-axis (Agency Name) and y-axis (count)
x = alt.X("Agency Name:N", title="Agency Name")  # Nominal for categorical data
y = alt.Y("count:Q", title="Number of Buildings")

# Build the chart with bars colored by Bldg Status
chart = chart.mark_bar().encode(
    x=x,
    y=y,
    color="Bldg Status:N",
    tooltip=["Agency Name", "Bldg Status", "count"]  # Add tooltip information
)

# Display the chart
chart
```

Explanation:

1. Import Libraries: I import pandas for data manipulation and Altair for building the chart.
2. Chart Base: I create a chart base using Chart(data).
3. Encoding Channels: I define the encoding channels for the x-axis (Agency Name) and y-axis (count).
    - alt.X("Agency Name:N", title="Agency Name"): This sets the x-axis to display the categorical data from the "Agency Name" column. "N" specifies nominal data type.
    - alt.Y("count:Q", title="Number of Buildings"): This sets the y-axis to display the quantitative data from the "count" column with the title "Number of Buildings". "Q" specifies quantitative data type.
4. Chart Building: I use mark_bar() to define the chart type as a bar chart. Then, we encode the x, y, and color channels on the chart.
    - color="Bldg Status:N": This assigns different colors to bars based on the values in the "Bldg Status" column.
5. Tooltip (Optional): I add a tooltip using the tooltip argument. This displays information about "Agency Name", "Bldg Status", and "count" when hovering over a bar.
6. Display Chart: Finally, we call chart.show() to display the visualization.
